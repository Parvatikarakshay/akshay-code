```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
iris_data=load_iris()

df=pd.DataFrame(iris_data.data,columns=['sepal length','sepal width','petal lengt','petal width'])
```

```python
df.head()
iris_data.data.shape
```

```python
x=iris_data.data
```

```python
y=iris_data.target.reshape(-1,1)#convert data into target data
```

```python
encoder=OneHotEncoder(sparse=False)
y=encoder.fit_transform(y)
```

```python
y
```

```python
#split the data for training and testing
tr_x,te_x,tr_y,te_y=train_test_split(x,y,test_size=0.20)
```

```python
model=Sequential()
```

```python
model.add(Dense(10,input_shape=(4,),activation ='linear',name='fc1'))#adding input layer and one hidden layer
model.add(Dense(3,activation='softmax',name='op'))
optimizer=Adam(lr=0.001)
model.compile(optimizer,loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
print (" network model summary")
model.summary()
```

```python
model.fit(tr_x,tr_y,verbose=2,batch_size=5,epochs=200)
result=model.evaluate (te_x,te_y)
print('final test set loss is' +   str(result[0]))
print ('final test set accuracy is' +str(result[1]))
```

```python
x_tra,x_val,y_tra,y_val=train_test_split(x,y,test_size=0.2)
```

```python
history=model.fit(x_tra,y_tra,epochs=200,batch_size=5,validation_data=(x_val,y_val))
```

```python
loss_train=history.history['loss']
loss_val=history.history['val_loss']
epochs=range(0,200)
plt.plot(epochs,loss_train,'g',label='Training loss')
plt.plot(epochs,loss_val,'b',label='Validation loss')
plt.title('Training and Validation loss')
plt.xlabel('Epochs')
plt.ylabel('loss')
plt.legend()
plt.show()
```

```python
loss_train=history.history['accuracy']
loss_val=history.history['val_accuracy']
epochs=range(0,200)
plt.plot(epochs,loss_train,'g',label='Training accuracy')
plt.plot(epochs,loss_val,'b',label='Validation accuracy')
plt.title('Training and Validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('accuracy')
plt.legend()
plt.show()
```

```python
y_pred=model.predict(te_x)
y_pred=(y_pred>0.5)
```
